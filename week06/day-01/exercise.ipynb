```python
#What are the ten most recently updated apps?


new_df = df2[df2['Last Updated'] != '1.0.19']

new_df['New Last Updated'] = list(map(lambda date: time.mktime(time.strptime(date,"%B %d, %Y")),new_df['Last Updated']))

new_df.nlargest(10, ['New Last Updated']) 





```

```python
#Find apps, that is in the top 25% of apps based on ratings and were updated in the last two months of dataset.

n = new_df['App'].count()
latest = new_df['New Last Updated'].max()

last_two_month = latest-(60*24*60*60)

tops = new_df.nlargest(int(n*0.25), ['Rating'])
tops[tops['New Last Updated'] >= last_two_month]

```

```python
#Which app not to download? - Find an app which takes the most space on your phone, but is rated poorly (define poorly)?
df3 = new_df[new_df['Size'] != 'Varies with device']

df3['Size']= (df3['Size'].replace(r'[kM]+$', '', regex=True).astype(float) * \
                 df3['Size'].str.extract(r'[\d\.]+([kM]+)', expand=False).fillna(1).replace(['k','M'], [10**3, 10**6]).astype(int))

df4 = df3.nlargest(10, ['Size'])
#pooly = rating < 4.0
df4[df4['Rating']< 4.0]
```

```python
#Create a reviews reliability of reviews index, which takes into account, how many reviews does the app have and how high the rating is. What metric do you take? Create as a new field.

def rating_reliability(rating):
    return rating < 5.0
     
def review_reliability(review):
    return review > 50

new_df['Is_Reliabilty'] = new_df['Rating'].apply(rating_reliability)&new_df['Reviews'].astype(int).apply(review_reliability)

    
```
