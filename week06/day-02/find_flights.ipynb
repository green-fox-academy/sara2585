```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python

url = 'https://api.skypicker.com/flights?flyFrom=BRS&to=PRG&dateFrom=18/12/2019&dateTo=23/12/2019&partner=picky&max_stopovers=0'
data = requests.get(url)
```

```python
df = json.loads(data.text)
```

```python
df1 = json_normalize(df['data'])
```

```python
df1.columns
```

```python
df1.sort_values(by=['price'], ascending = True)
```
