```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
data = pd.read_csv('metro-bike-share-trip-data.csv')
```

```python
new_data = data.dropna()
stops = new_data['Starting Station ID'].unique()
```

```python
times = new_data['Trip ID'].groupby(data['Starting Station ID']).count()
```

```python
#stops using frequency
plt.subplots(figsize= (20, 5))
plt.bar(x=stops,height = times)
plt.xticks(stops)
plt.title('using frequency by stop')
```

```python
data['Start Time'] = pd.to_datetime(data['Start Time'])
```

```python
data['End Time'] = pd.to_datetime(data['End Time'])
```

```python
hours = data['Start Time'].apply(lambda x: x.hour).unique()
```

```python
using_times = data['Trip ID'].groupby(data['Start Time'].apply(lambda x: x.hour)).count()
```

```python
#using frequency by hours
plt.subplots(figsize= (10, 5))
plt.bar(x=hours,height = using_times)
plt.xticks(hours)
plt.title('using frequency by hours')
```

```python
#using frequency by month
months = data['Start Time'].apply(lambda x: x.month).unique()
```

```python
using_times1 = data['Trip ID'].groupby(data['Start Time'].apply(lambda x: x.month)).count()
```

```python
#using frequency by month in a year
plt.subplots(figsize= (10, 5))
plt.bar(x=months,height = using_times1)
plt.xticks(months)
plt.title('using frequency by month')
```
